## Dataset Overview
This data set includes information about individual rides made in a bike-sharing system covering the greater San Francisco Bay area.

The dataset  contains 15 features of 171,230 bike trips. The included features includes:

* Trip Infomation: **duration(min),start_hour ,end_hour ,start_day, end_day, day_of_month, week_in_month,bike_id,  and distance(km)**

* Station Information: **start_station_name and  end_station_name**

* Member Infomation: **user_type, member_gender, bike_share_for_all_trip and member_age**


```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
#colours
base_color = sns.color_palette()

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df_master = pd.read_csv('df_master.csv')
```

## (Hourly Bike Usage)

It is observed that 8am and 5pm are the hour with the most activties indicating most of the trips are immediately before working hours 7am , in between working hours 10am to 3pm, and immediately after working hours 6pm to 7pm.


```python
fig, ax = (plt.subplots(nrows=1,ncols=2,figsize=(15,8)))
fig.suptitle('Trips start and end hour',fontsize=20)

plot1 = sns.countplot(data=df_master, x='start_hour', color=base_color[0],ax=ax[0]);
plot2 = sns.countplot(data=df_master, x='end_hour', color=base_color[5],ax=ax[1]);
```

## (Daily Bike Usage)


It is observed that majority of the trips are between Mondays and Fridays as expected since the hourly plot shows high activity during working hours with Satrudays and Sundays having a low number of trips


```python
day_order= ['Monday','Tuesday','Wednesday','Thursday','Friday','Sunday','Saturday']

fig, ax = (plt.subplots(nrows=1,ncols=2,figsize=(15,8)))
fig.suptitle('Trips Starting and Ending Day',fontsize=20)

plot1 = sns.countplot(x='start_day', data=df_master,color=base_color[0],order=day_order,ax=ax[0]);
plot2 = sns.countplot(x='end_day', data=df_master,color=base_color[5],order=day_order,ax=ax[1]);
```

## (How Each User Type Uses The Services Hourly )

We have seen how both users combined uses the service hourly but now we want to look at how both Subscribers and Customers uses it. 

Generally, Subscribers have the higher number of trips.
    
* **Start hour**
    8am and 5pm is the peak hour for both customers and subscribers
    
* **End hour**
    8am and 5pm is the peak hour for Subscribers  while 5pm and 6pm is the peak hour for Customers.


```python
fig, ax = (plt.subplots(2,figsize=(15,10)))
fig.suptitle('Hourly Trips by User Type For Both Starting And Ending Hour',fontsize=15)           
plot1 = sns.countplot(x='start_hour',hue='user_type',palette=["#ff7f0e",'#1f77b4'],data=df_master,ax=ax[0])
plot2= sns.countplot(x='end_hour',hue='user_type',palette=["#ff7f0e",'#1f77b4'],data=df_master,ax=ax[1])
```

## ( How Each User Type Uses The Services Daily )

Generally, Subscribers have a higher number of trips per day. Trips on weekdays(Mon- Fri) have a higher number of trips compared to the weekends(Sat-Sun) as expected. Although this only applies to Subscribers. With Customers having a fairly constant number of trips in a day
  
    

```python
day_order= ['Monday','Tuesday','Wednesday','Thursday','Friday','Sunday','Saturday']

fig, ax = (plt.subplots(nrows=1,ncols=2,figsize=(15,8)))
fig.suptitle('Daily Trips by User Types',fontsize=20)

plot1 = sns.countplot(x='start_day',hue='user_type', data=df_master,palette=["#ff7f0e",'#1f77b4'],order=day_order,ax=ax[0]);
plot2 = sns.countplot(x='end_day',hue='user_type', data=df_master,palette=["#ff7f0e",'#1f77b4'],order=day_order,ax=ax[1]);
```

## ( Who Spends More Time Biking And At What Hour?)

Female riders do spends more time on trips, having the highest duration for 22 hours of the 24 hours available. With 3am been the peak hour.


```python
fig, ax = (plt.subplots(nrows=2,ncols=1,figsize=(15,8)))
fig.suptitle('Trip Duration Starting And Ending Hour In A day by Male and Female Riders',fontsize=15)
plot1 = sns.pointplot(x='start_hour',y='duration(min)', ci =True,
    hue='member_gender',data=df_master, ax=ax[0])


plot2 = sns.pointplot(x='end_hour',y='duration(min)',ci =True,
    hue='member_gender', data=df_master, ax=ax[1])
```

## ( Who Spends More Time Biking And At    What Day?)
Again, Generally Female spends more time on trips and Sartuday happens to be the day with the most time spent on trips. Indicating trips on weekends tends to take more time

```python

fig, ax = (plt.subplots(nrows=1,ncols=2,figsize=(15,8)))
fig.suptitle('Trip Duration Starting And Ending Day By Male and Female Riders',fontsize=15)

plot1 = sns.pointplot(x="start_day",y='duration(min)',hue='member_gender', ci =True,
    data=df_master,palette=[base_color[0],base_color[1]],ax=ax[0],linestyles = '-',order=day_order)

plot2 = sns.pointplot(x="end_day",y='duration(min)',data=df_master,hue='member_gender',ci=True,
                     color=base_color[5],ax=ax[1],palette=[base_color[0],base_color[1]],linestyles = '-', order=day_order)
```
